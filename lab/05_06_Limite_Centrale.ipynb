# 05-06: Limite Centrale

```python
import numpy as np
from scipy.stats import norm, t
import matplotlib.pyplot as plt
```

```python
epochs = 1_000
threshold = 10

samples = []

for _ in range(epochs):

    count = 0
    summ = 0
    while summ <= threshold:
        
        summ += np.random.uniform(0,1)
        count += 1

        if summ >= threshold:
            samples.append(count)


samples = np.array(samples)
```

```python
mean = np.mean(samples)
std = np.std(samples, ddof=1)
n = samples.size
```

```python
std
```

```python
plt.hist(samples)
plt.show()
```

```python
x_gauss = np.linspace(mean-3*std,mean+3*std, 100)
y_gauss = norm.pdf(x_gauss, loc=mean, scale=std)
```

```python
x_t = np.linspace(mean-3*std,mean+3*std, 100)
y_t = t.pdf(x_t, n-1, loc=mean, scale=std)
```

```python
plt.hist(samples, density=True)
plt.plot(x_t, y_t, "y--")
plt.plot(x_gauss, y_gauss, "r--")
plt.show()
```
