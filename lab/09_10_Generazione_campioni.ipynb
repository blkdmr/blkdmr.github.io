## 09-10: Generazione campioni da distribuzione scelta

```python
from scipy.stats import multinomial, dirichlet
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Multinomiale

```python
n_cat = 4
supp = np.random.randint(0,n_cat, size=1000)
uniques, p = np.unique(supp,return_counts=True)
p = p / supp.size
p = np.round(p, 3)
print(p)
```

```python
n_trials = 20
gen = multinomial.rvs(n_trials, p, size=1000)
gen
```

```python
estimated_p = gen.mean(axis=0) / n_trials
estimated_p = np.round(estimated_p, 3)
print(estimated_p)
```
