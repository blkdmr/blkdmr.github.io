```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
```

## Normale teorica
Definiamo una distribuzione normale con media `mu=10` e deviazione standard `sigma=4`. Calcoliamo la funzione di densità (PDF) su un intervallo di valori.

```python
mean = 10
std = 4
x = np.linspace(mean-3*std, mean+3*std, 100)
y = norm.pdf(x, loc=mean, scale=std)

plt.plot(x,y, label="pdf")
plt.legend()
plt.show()
```

## Somma di variabili uniformi
Generiamo una matrice `samples` di dimensione `(1000, 1000)` dove ogni colonna è una sequenza di 1000 estrazioni da Uniforme(0,1).
Calcoliamo la somma **per colonna**.

```python
n = 1000
samples = np.random.uniform(0,1, size=(n,n))
print(samples.shape)
```

```python
summ = np.sum(samples, axis=0)
print(summ.shape)
mean = np.round(np.mean(summ),2)
std = np.round(np.std(summ),2)
print(mean)
print(std)
```

## Confronto con una normale stimata
Costruiamo una normale `N(mean_emp, std_emp)` e confrontiamo graficamente PDF teorica e istogramma delle somme.

```python
x = np.linspace(mean-3*std, mean+3*std, 100)
y = norm.pdf(x, loc=mean, scale=std)

plt.hist(summ, density=True)
plt.plot(x,y, "r--",label="theorical")
plt.show()
```
